```python
from autoscraper import AutoScraper
```

```python
amazon_url="https://www.amazon.in/s?k=iphones&crid=14HXJ1O132DEU&sprefix=iphones%2Caps%2C296&ref=nb_sb_noss"

wanted_list=["₹49,900", "Apple iPhone 11 (64GB) - Black"]
```

```python
scraper=AutoScraper()
result=scraper.build(amazon_url,wanted_list)
print(result)
```

```python
scraper.get_result_similar(amazon_url, grouped=True)
```

```python
scraper.set_rule_aliases({'rule_5ljl':'Title', 'rule_ewat':'Price'})
scraper.keep_rules(['rule_5ljl', 'rule_ewat'])
scraper.save('amazon-search')
```

```python
result=scraper.get_result_similar('https://www.amazon.in/s?k=mi+phones&crid=3FAVCGF4KHLF3&sprefix=mi+phones%2Caps%2C262&ref=nb_sb_noss', group_by_alias=True)
```
